# *Handling Null Values - Imputation*

```python
df1['normalized-losses'].mean() #no outliers
```

```python
df1['normalized-losses'].median() #if outliers exist
```

```python
mean = df1['normalized-losses'].mean()
mean
```

```python
df1['normalized-losses'] =df1['normalized-losses'].fillna(mean)
```
