# *Predicting Car Price Using KNN Algorithm*

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor # NEW IMPORT for KNN
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics import r2_score
```

```python
df2.head()
```

```python
# Replace non-numeric values ('?') with NaN
df2['horsepower'] = df2['horsepower'].replace('?', np.nan)
```

```python
# Convert columns to numeric, forcing errors to NaN
df2['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
```

```python
# Drop rows where any of the critical features or price are missing
df2.dropna(subset=['horsepower', 'engine-size', 'curb-weight', 'price'], inplace=True)
```

```python
df2.info()
```

```python
df2.isnull().sum()
```

```python
df2_x= df2[['curb-weight', 'horsepower', 'engine-size']]
df2_y = df2['price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(df2_x, df2_y, test_size=0.2, random_state=42)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
len(y_train)
```

```python
len(y_test)
```

```python
# Initialize the KNN Regressor (using k=5 neighbors)
knn_reg = KNeighborsRegressor(n_neighbors=5)
```

```python
# Fit the KNN model using the training data
knn_reg.fit(x_train,y_train)
```

```python
# Predict prices using the test set
y_predict_knn = knn_reg.predict(x_test)
```

```python
y_predict_knn
```

```python
# Evaluate the model's performance using R-squared
score_knn = r2_score(y_test, y_predict_knn)
```

```python
score_knn #(using k=5 neighbors)
```

```python
# Initialize the KNN Regressor (using k=6 neighbors)
knn_reg = KNeighborsRegressor(n_neighbors=6)
```

```python
# Fit the KNN model using the training data
knn_reg.fit(x_train,y_train)
```

```python
# Predict prices using the test set
y_predict_knn = knn_reg.predict(x_test)
```

```python
y_predict_knn
```

```python
# Evaluate the model's performance using R-squared
score_knn = r2_score(y_test, y_predict_knn)
```

```python
score_knn   #(using k=6 neighbors)
```
